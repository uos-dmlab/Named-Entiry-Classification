```python
max_len = 100
bs = 32
p_index_to_ner={0: 'SS', 1: 'NNP', 2: 'SP', 3: 'NNG', 4: 'VV', 5: 'EC', 6: 'SN', 7: 'NNB', 8: 'EP', 9: 'EF', 10: 'SL', 11: 'JKB', 12: 'XSV', 13: 'ETM', 14: 'JKS', 15: 'JKO', 16: 'SF', 17: 'JX', 18: 'JKQ', 19: 'JC', 20: 'MAG', 21: 'JKG', 22: 'XSN', 23: 'NP', 24: 'VCP', 25: 'MMN', 26: 'VA', 27: 'SO', 28: 'SW', 29: 'VX', 30: 'NR', 31: 'ETN', 32: 'XPN', 33: 'MMD', 34: 'SH', 35: 'XSA', 36: 'MAJ', 37: 'SE', 38: 'MMA', 39: 'JKC', 40: 'VCN', 41: 'NA', 42: 'XR', 43: 'IC', 44: 'NF', 45: 'JKV', 46: 'NV', 47: '[PAD]'}
index_to_ner={0: 'PER_B', 1: 'DAT_B', 2: '-', 3: 'ORG_B', 4: 'CVL_B', 5: 'NUM_B', 6: 'LOC_B', 7: 'EVT_B', 8: 'TRM_B', 9: 'TRM_I', 10: 'EVT_I', 11: 'PER_I', 12: 'CVL_I', 13: 'NUM_I', 14: 'TIM_B', 15: 'TIM_I', 16: 'ORG_I', 17: 'DAT_I', 18: 'ANM_B', 19: 'MAT_B', 20: 'MAT_I', 21: 'AFW_B', 22: 'FLD_B', 23: 'LOC_I', 24: 'AFW_I', 25: 'PLT_B', 26: 'FLD_I', 27: 'ANM_I', 28: 'PLT_I', 29: '[PAD]'}
def inference(test_sentence):
    ret_token=[]
    i=-1
    p_tokenized_sentence = np.array([tokenizer.encode(test_sentence, max_length=max_len, truncation=True, padding='max_length')])
    p_tokenized_mask = np.array([[int(x!=1) for x in p_tokenized_sentence[0].tolist()]])
    p_ans = p_nr_model.predict([p_tokenized_sentence, p_tokenized_mask])
    p_ans = np.argmax(p_ans, axis=2)

    tokenized_sentence = np.array([tokenizer.encode(test_sentence, max_length=max_len, truncation=True, padding='max_length')])
    tokenized_mask = np.array([[int(x!=1) for x in tokenized_sentence[0].tolist()]])
    ans = nr_model.predict([tokenized_sentence, tokenized_mask])
    ans = np.argmax(ans, axis=2)

    p_tokens = tokenizer.convert_ids_to_tokens(p_tokenized_sentence[0])
    new_tokens, new_p_labels, new_n_labels = [], [],[]
    for token, p_label_idx, n_label_idx in zip(p_tokens, p_ans[0],ans[0]):
        if (token.startswith("##")):
            #new_labels.append(p_index_to_ner[label_idx])
            new_tokens[i]=new_tokens[i]+token[2:]
        elif (token=='[CLS]'):
            pass
        elif (token=='[SEP]'):
            pass
        elif (token=='[PAD]'):
            pass
        elif (token != '[CLS]' or token != '[SEP]'):
            new_tokens.append(token)
            new_p_labels.append(p_index_to_ner[p_label_idx])
            new_n_labels.append(index_to_ner[n_label_idx])
            i=i+1
    for num in range(len(new_tokens)):
        #print(new_tokens[num],"(",new_p_labels[num],",",new_n_labels[num],")")
        ret_token.append([new_tokens[num],new_p_labels[num],new_n_labels[num]])
    return ret_token
    
```

```python
#기존에 만들어놓은 인명,기관명,지명을 사용하는 코드
Per_list=[]
Org_list=[]
Loc_list=[]
Per_csv=pd.read_csv('/content/drive/MyDrive/NLP/Per_df')
Org_csv=pd.read_csv('/content/drive/MyDrive/NLP/Org_df')
Loc_csv=pd.read_csv('/content/drive/MyDrive/NLP/Loc_df')

for i in range(len(Per_csv['Word'])):
    Per_list.append(Per_csv['Word'][i])
for i in range(len(Org_csv['Word'])):
    Org_list.append(Org_csv['Word'][i])
for i in range(len(Loc_csv['Word'])):
    Loc_list.append(Loc_csv['Word'][i])
```

```python
voc_per=[ word for word in Per_list if word in n2vmodel]
voc_org=[ word for word in Org_list if word in n2vmodel]
voc_loc=[ word for word in Loc_list if word in n2vmodel]
voc_per=list(set(voc_per))
voc_org=list(set(voc_org))
voc_loc=list(set(voc_loc))
```

```python
def ranking(per,my_word):
    cos_input=n2vmodel[my_word]

    per_vector = []
    for word in per:
        per_vector.append(n2vmodel[word])
    zero_data = np.zeros(shape=(1,len(per)))
    per_array = pd.DataFrame(zero_data, columns=per)
    for i in range(len(per)):
        per_array.iloc[0,i]=cos_sim(cos_input,per_vector[i])
    per_array=(per_array.loc[0,:].sort_values(ascending=False))
    return(per_array[:5])
```

```python
def rank_all(per,org,loc):
    input_word=input()
    print("Per_Ranking","*"*20)
    ranking(per,input_word)
    print("Org_Ranking","*"*20)
    ranking(org,input_word)
    print("Loc_Ranking","*"*20)
    ranking(loc,input_word)
```
